```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn import tree
from scipy.stats import mode
```

```python
def treeCrossValidate(data, model, name=''):
    data['train'] = (np.random.uniform(0,1, len(data)) * 10).astype(int)
    err = 0
    print(f"\n*** {name} ***")
    for i in range(10):
        data_train = data[data['train'] != i]
        data_test = data[data['train'] == i]
        
        y_train = data_train['class']
        X_train = data_train.drop('class', axis = 1)
        model.fit(X_train, y_train)
        y_test = data_test['class']
        X_test = data_test.drop('class', axis = 1)
        
        err += model.score(X_test, y_test)
#         print(f'\t{model.score(X_test, y_test)}')
    print('Score: ', err/10)
    return err
```

```python
features = pd.read_csv('data/7/features.txt', delim_whitespace=True, header=None)
train_data = pd.read_csv('data/7/train/X_train.txt', delim_whitespace=True, names=features[1].ravel())
train_data['class'] = pd.read_csv('data/7/train/y_train.txt', delim_whitespace=True, header=None)
test_data = pd.read_csv('data/7/test/X_test.txt', delim_whitespace=True, names=features[1].ravel())
test_data['class'] = pd.read_csv('data/7/test/y_test.txt', delim_whitespace=True, header=None)

data = train_data + test_data
data.head()
```

```python
for name in data.columns.values:
    if data[name].dtype == 'object':
        print(name)
    data[name] = pd.Categorical(data[name]).codes
data.head()
```

```python
models = [
    (
        RandomForestClassifier(n_estimators=20, max_depth=13, min_samples_split=10),
        'RandomForestClassifier'
    ),
    (
        ExtraTreesClassifier(n_estimators=30, min_samples_split=20, max_leaf_nodes=15),
        'ExtraTreesClassifier'
    ),
    (
        tree.DecisionTreeClassifier(), 'Default DecisionTreeClassifier'
    ),
    (
        RandomForestClassifier(), 'Default RandomForestClassifier'
    )
]
```

```python
for model, name in models:
    treeCrossValidate(data, model, name)
```

```python
def knn(data, labels, test, k):
    a = np.sum(test ** 2, axis=-1)
    b = np.sum(data ** 2, axis=-1)
    c = -2 * (test @ data.T)
    dists = c + a.reshape(-1,1) + b
    knn = np.argpartition(dists, k)[..., :k]
    knn_labels = labels[knn]
    res = mode(knn_labels, axis=1)[0]
    return res[:,0]
```

```python
def treeCrossValidateKNN(data, k):
    data['train'] = (np.random.uniform(0,1, len(data)) * 10).astype(int)
    err = 0
    print(f"\n*** k = {k} ***")
    for i in range(10):
        data_train = data[data['train'] != i]
        data_test = data[data['train'] == i]
        
        y_train = data_train['class']
        X_train = data_train.drop('class', axis = 1)
        y_test = data_test['class']
        X_test = data_test.drop('class', axis = 1)
        
        knn_test_labels = knn(X_train.as_matrix(), y_train.as_matrix(), X_test.as_matrix(), k)
        
        err += np.sum(knn_test_labels == y_test)/len(y_test)
        print(f'\t{ np.sum(knn_test_labels == y_test)/len(y_test)}')
    print('Score: ', err/10)
    return err/10
```

```python
for k in range(3,8):    
    s = treeCrossValidateKNN(data, k)
```
